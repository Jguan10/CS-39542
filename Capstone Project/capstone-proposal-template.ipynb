```python
import pandas as pd
import seaborn as sns

df1 = pd.read_csv('anime.csv')
df2 = pd.read_csv('animes.csv')
df3 = pd.read_csv('anime-dataset-2023.csv')
```

```python
df2.info()
```

```python
df2 = df2.drop(['img_url', 'link'], axis = 1)
```

```python
df2.info()
```

```python
# Combine df1 and df2 on id

df2 = df2.rename(columns = {'uid': 'anime_id'})
merged_df = pd.merge(df1, df2, on = 'anime_id', how = 'inner')
merged_df.drop_duplicates(keep = 'first', inplace = True)
merged_df.head(2)
```

```python
merged_df.info()
```

```python
merged_df = merged_df.drop(['episodes_x', 'members_x', 'genre_y'], axis = 1)
merged_df.info()
```

```python
merged_df = merged_df.dropna()
merged_df = merged_df.rename(columns = {'genre_x': 'Genre',
                                        'episodes_y' : 'Episodes',
                                        'members_y' : 'Members'})
merged_df.info()
```

```python
df3.head()
```

```python
df3.info()
```

```python
df3 = df3.drop(['Premiered', 'Image URL', 'Scored By'], axis = 1)
df3.info()
```

```python
final_df = pd.merge(merged_df, df3, on = 'anime_id', how = 'outer')
final_df.info()
```

```python
final_df = final_df.drop_duplicates(subset = ['anime_id'], keep = 'first')
final_df.info()
```

```python
import pandas as pd

df = pd.read_csv('Data/jikan_final.csv')
df.head(10)
```

```python
df.info()
```

```python
df = df[df['approved'] != False]
df.info()
```

```python
df = df.drop(columns = ['approved', 'scored_by', 'url', 'images', 'trailer', 
                        'titles', 'title_synonyms', 'title_japanese', 
                        'demographics', 'year', 'season', 'airing', 'status', 
                        'duration', 'explicit_genres']) 
```

```python
df.info()
```

```python
theme_view = df[['themes']]
theme_view
```

```python
df = df.drop(columns = ['themes', 'background', 'favorites', 'members'])
```

```python
df.info()
```

```python
df.head(1)
```

```python
df = df.drop(columns = ['aired', 'broadcast'])
df.info()
```

```python
df = df.rename(columns={"mal_id": "anime_id"})
```

Further data manipulation, hot encoding and preparing the features of synopsis, rating, source, and genres

```python
import pandas as pd 
import ast
```

```python
df = pd.read_csv('Data/nlp.csv')
df.head(10)
```

```python
df.info()
```

Dropping producers, studios, and licensors because there are way too many empty values

```python
df = df[['anime_id', 'title', 'episodes','source','type', 'synopsis', 'genres', 'score', 'Popularity_category', 'Rank_category']]
df.info()
```

```python
df.head(10)
```

Hot encode the genres

```python
def extract_unique_values(df, column_name):
    unique_values = set() 
    for row in df[column_name]:
        if isinstance(row, list): 
            unique_values.update(row)
    return list(unique_values)

df["genres"] = df["genres"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
test = extract_unique_values(df, "genres")
test
```

```python
df = df.explode("genres")
df = pd.get_dummies(df, columns=["genres"], prefix="Genres", prefix_sep="_")
df.info()
```

Hot encode the types

```python
df = df.drop_duplicates(subset="anime_id", keep="first").reset_index(drop=True)
df.info()
```

```python
df['type'].value_counts()
```

```python
df = pd.get_dummies(df, columns=["type"], prefix="Types", prefix_sep="_")
df.info()
```

Hot encode source

```python
df['source'].value_counts()
```

```python
import numpy as np

df['source'] = df['source'].replace('Unknown', np.nan)
df['source'] = df['source'].replace('Mixed media', np.nan)
df['source'] = df['source'].replace('Radio', np.nan)
df['source'] = df['source'].replace('Card game', 'Game')
df['source'] = df['source'].replace('Picture book', 'Book')
df['source'] = df['source'].replace('Other', np.nan)
df['source'] = df['source'].replace('Web manga', 'Manga')
df['source'] = df['source'].replace('4-koma manga', 'Manga')
df['source'] = df['source'].replace('Music', np.nan)
df['source'] = df['source'].replace('Web novel', 'Book')
df['source'] = df['source'].replace('Novel', 'Book')
df['source'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['source'], prefix='Source', prefix_sep='_')
df.info()
```

```python
df.to_csv('genres_synopsis_type.csv', index = False)
```
